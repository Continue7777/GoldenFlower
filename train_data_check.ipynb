```python
import pandas as pd
```

```python
df = pd.read_csv("train_data.csv",sep='\t',header=None,names=["observation","action","reward","done","next_observation","cardB","action_status","nowprice"])
```

```python
import GameEnv
gameEnv = GameEnv.GlodenFlower([2000,2000])
d = {"豹子":10,"同花顺":9,"金花":8,"顺子":7,"对子":6,"单":5}
reverse_dict = {v: k for k, v in d.iteritems()}
def f(x,i):
    return eval(x)[i]
def f1(x):
    try:
        return reverse_dict[gameEnv.score(eval(x)[1])]
    except:
        return "闷"
    
def f2(x):
    try:
        return reverse_dict[gameEnv.score(eval(x))]
    except:
        return "闷"
    
df["seq"] = df['observation'].apply(lambda x:f(x,0))
df["next_seq"] = df['next_observation'].apply(lambda x:f(x,0))
df['next_seq_len'] = df['next_seq'].apply(lambda x:len(x))
df["cardA"] = df['observation'].apply(lambda x:f(x,1))
df['status'] = df['observation'].apply(lambda x:f(x,2))
df["card_levelA"] = df['observation'].apply(lambda x:f1(x))
df["card_levelB"] = df['cardB'].apply(lambda x:f2(x))
```
